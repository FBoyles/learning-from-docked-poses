```python
import pandas as pd
xtal = pd.Series({'Vina': 0.574, 'Structure': 0.765, 'Structure + Ligand': 0.789})
dock = pd.Series({'Vina': 0.504, 'Structure': 0.706, 'Structure + Ligand': 0.766})
```

```python
from matplotlib import pyplot as plt


plt.style.use('fivethirtyeight')

plt.rcParams['axes.facecolor']='white'
plt.rcParams['axes.edgecolor']='white'
plt.rcParams['figure.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,6))
ax.set_xlabel('Model')
ax.set_ylabel('Pearson Correlation Coefficient')
xtal.plot.bar(ax=ax, color=['red', 'green', 'green'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.set_ylim(0.4, 0.8)
fig.savefig('/home/fergus/Desktop/xtal_plot.png')
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,6))
ax.set_xlabel('Model')
ax.set_ylabel('Pearson Correlation Coefficient')
dock.plot.bar(ax=ax, color=['red', 'orange', 'green'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.set_ylim(0.4, 0.8)
fig.savefig('/home/fergus/Desktop/dock_plot.png')
```
